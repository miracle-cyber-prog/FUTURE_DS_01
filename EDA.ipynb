```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("sales.csv")
```

### Revenue Trend Over Time

```python
monthly_sales = df.groupby(["Year", "Month"])["Sales"].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x="Month", y = "Sales", hue ="Year", marker ="o")
plt.title("Monthly Sales Trend by year")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

### Top_Selling Products

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y = top_products.index, palette="mako")
plt.title("Top 10 Selling Products")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()
```

### Category and Subcategory Analysis

```python
category_sales = df.groupby(["Category", "Sub-Category"])["Sales"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data = category_sales, x="Sub-Category", y = "Sales", hue = "Category")
plt.title("Sales by Category and Subcategory")
plt.xticks(rotation=45)
plt.show()
```

### regional Performance

```python
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=state_sales.index, y =state_sales.values, palette="viridis")
plt.title("Top States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### profit analysis

```python
profit_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=profit_category.index, y =profit_category.values, palette="coolwarm")
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.show()
```

### Profit Margin Distribution

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = df, x="Category", y ="Profit", palette="Set2")
plt.title("Profit margin Distribution by Category")
plt.xticks(rotation=45)
plt.show()
```
