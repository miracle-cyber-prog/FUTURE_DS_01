```python
df.drop_duplicates(inplace = True)

#covert date
df["Order Date"] = pd.to_datetime(df["Order Date"])

#ensure month and year column exist
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year


#Feature Engineering: profit margin
df["Profit Margin"] = df["Profit"] / df["Sales"]


#flag top performing produacts (sales > 90 percent)
top_sales_thresholds = df["Sales"].quantile(0.9)
df["Top products"] = np.where(df["Sales"] >= top_sales_thresholds, "Yes", "No")
```
